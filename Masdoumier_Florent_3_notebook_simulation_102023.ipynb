## Fonctions utils

```python
from scipy.spatial import distance_matrix

def relabel_clusters_based_on_centroids(old_centroids, current_labels, normalized_data):
    new_centroids = np.array([normalized_data[current_labels == i].mean(axis=0) for i in range(old_centroids.shape[0])])
    distances = distance_matrix(new_centroids, old_centroids)
    mapping = np.argmin(distances, axis=1)
    relabeled_clusters = np.array([mapping[label] for label in current_labels])
    return relabeled_clusters, new_centroids
```

## Segmentation des donnÃ©es par mois

```python
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
df['Month'] = df['order_purchase_timestamp'].dt.to_period('M')

latest_date = df['order_purchase_timestamp'].max()
start_date = latest_date - pd.DateOffset(months=12)
df = df[(df['order_purchase_timestamp'] >= start_date) & (df['order_purchase_timestamp'] <= latest_date)]

df.to_csv('rfm_by_month.csv', index=False)
```

```python
# Lister et supprimer les mois qui ont 5 lignes ou moins, il ne faut pas un nombre de samples < au nombre de clsuters
month_counts = df['Month'].value_counts()
months_to_drop = month_counts[month_counts <= 5].index
print(months_to_drop)
df = df[~df['Month'].isin(months_to_drop)]
```

### Normaliser les donnÃ©es

```python
from sklearn.metrics import silhouette_score
from sklearn.metrics import davies_bouldin_score

cost = []
silhouettes = []
dbis = []
K = range(2, 10)

reference_date = df['order_purchase_timestamp'].max() + pd.Timedelta(days=1)
df['Recency'] = (reference_date - df['order_purchase_timestamp']).dt.days

rfm = df.groupby('customer_unique_id').agg({
    'Recency': 'min',
    'order_purchase_timestamp': 'count',
    'total_price_per_order': 'sum',
    'review_score': 'mean',
    'payment_type_credit_card': 'sum',
}).rename(columns={
    'order_purchase_timestamp': 'Frequency',
    'total_price_per_order': 'Monetary',
    'review_score': 'Review_score_Mean'
})
rfm = rfm.reset_index()

scaler = StandardScaler()
numeric_columns = rfm.select_dtypes(include=[np.number])
rfm_normalized = scaler.fit_transform(numeric_columns)
```

```python
for k in K:
    kmeans = KMeans(n_clusters=k, n_init=100, init='k-means++', max_iter=200)
    kmeans.fit(rfm_normalized)
    cost.append(kmeans.inertia_)
    silhouettes.append(silhouette_score(rfm_normalized, kmeans.labels_))
    dbis.append(davies_bouldin_score(rfm_normalized, kmeans.labels_))

plt.plot(range(2, 10), cost, marker='o')
plt.xlabel('Nombre de clusters')
plt.ylabel('CoÃ»t (Inertie)')
plt.title('MÃ©thode du coude pour le choix optimal de k')
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(K, silhouettes, 'bx-')
plt.xlabel('k')
plt.ylabel('Score de silhouette')
plt.title('Le score de silhouette montrant le k optimal')
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(K, dbis, 'bx-')
plt.xlabel('k')
plt.ylabel('Score de Davies Bouldin')
plt.title('Le score de Davies Bouldin par k')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import adjusted_rand_score
from datetime import datetime
from dateutil.relativedelta import relativedelta

end_period = df['Month'].max()
end_period_str = end_period.strftime('%Y-%m')
end_period_date = datetime.strptime(end_period_str, '%Y-%m')
start_period_date = end_period_date - relativedelta(months=11)
start_period_str = start_period_date.strftime('%Y-%m')

silhouette_scores = []
intra_cluster_variance = []
inter_cluster_variance = []
ari_scores = []

previous_clusters = None
previous_ids = None  # Pour stocker les ID clients de la pÃ©riode prÃ©cÃ©dente

start_period_date += pd.DateOffset(months=1)
difference_in_months = relativedelta(end_period_date, start_period_date).months

for period_offset in range(1, difference_in_months + 3):
    if period_offset == 1:
        end_period = df['Month'].max()
        end_period_str = end_period.strftime('%Y-%m')
        end_period_date = datetime.strptime(end_period_str, '%Y-%m')
        end_period_date = end_period_date - relativedelta(months=11)
        start_period_date = end_period_date - relativedelta(months=12)
        start_period_str = start_period_date.strftime('%Y-%m')
        end_of_period_date = end_period_date
        end_of_period_period = pd.Period(end_of_period_date, freq='M')
        period_data = df_reference.copy()

    if period_offset > 1:
        end_period = df['Month'].max()
        end_period_str = end_period.strftime('%Y-%m')
        end_period_date = datetime.strptime(end_period_str, '%Y-%m')
        start_period_date = end_period_date - relativedelta(months=11)
        start_period_str = start_period_date.strftime('%Y-%m')
        new_period_offset = period_offset - 1
        end_of_period_date = start_period_date + pd.DateOffset(months=new_period_offset)
        end_of_period_period = pd.Period(end_of_period_date, freq='M')
        print(end_of_period_date)
        print(end_of_period_period)
        period_data = df[df['Month'] < end_of_period_period].copy()


    print("\n\nPeriod start date :", start_period_date.strftime('%Y-%m-%d'))
    print("Period end date :", end_of_period_date.strftime('%Y-%m-%d'))
    print("Periode max", period_data['order_purchase_timestamp'].max())

    reference_date = period_data['order_purchase_timestamp'].max() + pd.Timedelta(days=1)
    period_data['Recency'] = (reference_date - period_data['order_purchase_timestamp']).dt.days
    period_data['Delivery_wait_days'] = (pd.to_datetime(period_data['order_delivered_customer_date']) - pd.to_datetime(period_data['order_purchase_timestamp'])).dt.days


    rfm = period_data.groupby('customer_unique_id').agg({
        'Recency': 'min',
        'order_purchase_timestamp': 'count',
        'total_price_per_order': 'sum',
        'review_score': 'mean',
        'payment_type_credit_card': 'sum'
    }).rename(columns={
        'order_purchase_timestamp': 'Frequency',
        'total_price_per_order': 'Monetary',
        'review_score': 'Review_score_mean'
    })

    rfm = rfm.reset_index()
    rfm_filename = "rfm_monthly_" + str(period_offset) + ".csv"
    rfm.to_csv(rfm_filename, index=False)
    print(rfm.shape)
    numeric_columns = rfm.select_dtypes(include=[np.number])

    if period_offset == 1:
        # Normalisation
        scaler = StandardScaler()
        period_data_normalized = scaler.fit_transform(numeric_columns)
        
        previous_kmeans = KMeans(n_clusters=5, n_init=100, init='k-means++', max_iter=200)
        previous_fit = previous_kmeans.fit(period_data_normalized)
        previous_predict = previous_kmeans.predict(period_data_normalized)
        rfm_previous = rfm.copy()
        rfm_previous['Cluster'] = previous_predict
        previous_ids = set(rfm_previous['customer_unique_id'])
        previous_clusters = rfm_previous['Cluster'].to_numpy().copy()
    
    if period_offset > 1:
        period_data_normalized = scaler.transform(numeric_columns)
        previous_predict = previous_kmeans.predict(period_data_normalized)
        rfm_previous = rfm.copy()
        rfm_previous['Cluster'] = previous_predict
        previous_clusters = rfm_previous['Cluster'].to_numpy().copy()
        previous_ids = set(rfm_previous['customer_unique_id'])

    # Clustering
    kmeans = KMeans(n_clusters=5, n_init=100, init='k-means++', max_iter=200)
    current_fit = kmeans.fit(period_data_normalized)
    current_predict = kmeans.predict(period_data_normalized)

   # old_centroids = previous_kmeans.cluster_centers_
    #relabeled_clusters, new_centroids = relabel_clusters_based_on_centroids(old_centroids, current_predict, period_data_normalized)

    rfm['Cluster'] = current_predict

    # Calcul de l'ARI
    common_ids = set(rfm['customer_unique_id'])
    if previous_ids:
        common_ids = common_ids.intersection(previous_ids)
        common_clusters_current = rfm[rfm['customer_unique_id'].isin(common_ids)]['Cluster'].to_numpy()
        if previous_clusters is not None:
            common_index = rfm_previous['customer_unique_id'].isin(common_ids)
            common_clusters_previous = previous_clusters[np.where(common_index)]
            if (period_offset == 1):
                differences = [(i, a, b) for i, (a, b) in enumerate(zip(common_clusters_previous, common_clusters_current)) if a != b]
                #print(differences)
            ari_scores.append(adjusted_rand_score(common_clusters_previous, common_clusters_current))
        else:
            ari_scores.append(None)
```
